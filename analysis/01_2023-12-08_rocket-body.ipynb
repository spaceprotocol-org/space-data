## What does this notebook do?
This notebook accesses data from [space-track.org](https://www.space-track.org/) to analyse the lifespan of rocket bodies in orbit. The data is a part of space-track's [gp](https://www.space-track.org/documentation#api-basicSpaceDataGp) class ( general perturbations ), which is an efficient listing of the newest SGP4 keplerian element set for each artificial earth-orbiting object tracked by the 18th Space Defense Squadron.

```python
import io
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
username = None
password = None

if not username or not password:
    print("Enter valid credentials")

siteCred = { 'identity': username, 'password': password }
```

```python
# Error class to handle 'POST' and 'GET' errors
class HandleError(Exception):
    def __init__(self, args):
        Exception.__init__(self, "Exception raised with args {0}".format(args))
        self.args = args
```

```python
uriBase = "https://www.space-track.org"
requestLogin = "/ajaxauth/login"
requestQuery = "/basicspacedata/query"

# OBJECT_TYPE = ROCKET BODY
rocketBody = "/class/gp/OBJECT_TYPE/ROCKET%20BODY"

# Filters data by LAUNCH_DATE
predicate1  = "/LAUNCH_DATE/%3C%3Enull"

# Filters data by DECAY_DATE
predicate2 = "/DECAY_DATE/%3C%3Enull"

# Orders the data by NORAD_CAT_ID
orderBy = "/orderby/NORAD_CAT_ID%20asc"
```

```python
with requests.Session() as session:
    postCheck = session.post(uriBase + requestLogin, data = siteCred)
    if postCheck.status_code != 200:
        raise HandleError('POST')
    
    # Requests for ROCKET BODY data where LAUNCH_DATE and DECAY_DATE != null
    response1 = session.get(uriBase + requestQuery + rocketBody + predicate1 + predicate2 + orderBy)
    if response1.status_code != 200:
        raise HandleError('GET')
        
    data1 = response1.text
        
    dataFrame1 = pd.read_json(io.StringIO(data1))    
    
    """
    Using any available fonts is permissible for the plots, 
    but I chose CMU Serif for aesthetic reasons.
    For Ubuntu, you can install the font package fonts-cmu 
    (using sudo apt install fonts-cmu), which contains the fonts.
    """
    # plt.rcParams['font.family'] = "CMU Serif"
    # plt.rcParams['font.size'] = 12

    dataFrame1['LAUNCH_DATE'] = pd.to_datetime(dataFrame1['LAUNCH_DATE'])
    dataFrame1['DECAY_DATE'] = pd.to_datetime(dataFrame1['DECAY_DATE'])
    
    dataFrame1['LIFESPAN'] = ((dataFrame1['DECAY_DATE'] - dataFrame1['LAUNCH_DATE']).dt.days) / 365.25
    
    # Plots the mean lifespan of rocket bodies
    dataFrame1.groupby(dataFrame1['LAUNCH_DATE'].dt.year)['LIFESPAN'].mean().plot()
    
    plt.xlabel("Launch year")
    plt.ylabel("Mean LIFESPAN (years)")
    plt.title("LIFESPAN trend of rocket bodies")

session.close()
```
